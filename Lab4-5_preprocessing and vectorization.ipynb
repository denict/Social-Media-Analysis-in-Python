```python
punct = list(punctuation) # –°–ø–∏—Å–æ–∫ —Å–æ –∑–Ω–∞–∫–∞–º–∏ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è
punct.extend(['üáπüá∑', '¬´', '¬ª', '‚Äî', '*'])
punct.extend([])
punct += list(emoji.EMOJI_DATA.keys())
```

```python
df
```

```python
def del_punc(st, p):
  st_copy = st.copy()
  for char in p:
      while char in st_copy:
          st_copy.remove(char)
  return st_copy

def lowerr(st):
  for i in range(len(st)):
      st[i] = st[i].lower()
  return st

df['update_text'] = df['update_text'].apply(del_punc, args=[punct])
df['update_text'] = df['update_text'].apply(lowerr)
```

```python
df["update_text"][0]
```

```python
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words('russian')

def del_banwords(st, p):
  st_copy = st.copy()
  for char in p:
      while char in st_copy:
          st_copy.remove(char)
  return st_copy

df['update_text'] = df['update_text'].apply(del_banwords, args=[stopwords])
```

```python
morph = MorphAnalyzer()

def get_normis(st):
    return [morph.parse(word)[0].normal_form for word in st]

df['update_text'] = df['update_text'].apply(get_normis)
```

```python
# import re
# def clean_word(word):
#     return re.sub(r'[^–∞-—è–ê-–Ø—ë–Å]', '', word)

# def filter_russian(text):
#     words = str(text).split()
#     cleaned_words = [clean_word(word) for word in words]
#     return [word for word in cleaned_words if word and re.fullmatch(r'[–∞-—è–ê-–Ø—ë–Å]+', word)]

# df1['update_text'] = df1['update_text'].apply(filter_russian)
# df1['update_text'] = df1['update_text'].apply(' '.join)
import re

# –®–∞–≥ 1: –£–¥–∞–ª–µ–Ω–∏–µ —Å—Å—ã–ª–æ–∫ –∏–∑ –≤—Å–µ–≥–æ —Ç–µ–∫—Å—Ç–∞
def remove_links(text):
    # –ü–∞—Ç—Ç–µ—Ä–Ω –¥–ª—è —Å—Å—ã–ª–æ–∫: http(s), www, telegram-—Å—Å—ã–ª–∫–∏, –¥–æ–º–µ–Ω—ã
    url_pattern = r'(https?://\S+|www\.\S+|//t\.me/\S+|\S+\.(com|ru|net|org)/\S*)'
    return re.sub(url_pattern, '', str(text))

# –®–∞–≥ 2: –°—É—â–µ—Å—Ç–≤—É—é—â–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –æ—á–∏—Å—Ç–∫–∏ (–º–æ–¥–∏—Ñ–∏—Ü–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ)
def clean_word(word):
    return re.sub(r'[^–∞-—è–ê-–Ø—ë–Å]', '', word)

def filter_russian(text):
    words = str(text).split()
    cleaned_words = [clean_word(word) for word in words]
    return [word for word in cleaned_words if word and re.fullmatch(r'[–∞-—è–ê-–Ø—ë–Å]+', word)]

# –ü—Ä–∏–º–µ–Ω—è–µ–º –æ–±—Ä–∞–±–æ—Ç–∫—É
df1['update_text'] = df1['update_text'].apply(remove_links)  # –°–Ω–∞—á–∞–ª–∞ —É–¥–∞–ª—è–µ–º —Å—Å—ã–ª–∫–∏
df['update_text'] = df['update_text'].apply(filter_russian)
df['update_text'] = df['update_text'].apply(lambda x: ' '.join(x) if x else '')
```

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df2['update_text'] = df2['update_text'].fillna('')
# –ó–∞–º–µ–Ω–∏—Ç—å –ø—É—Å—Ç—ã–µ —Å—Ç—Ä–æ–∫–∏ –∏ NaN –Ω–∞ –º–µ—Ç–∫—É [EMPTY]
df2['update_text'] = df2['update_text'].fillna('[EMPTY]')
df2['update_text'] = df2['update_text'].replace(r'^\s*$', '[EMPTY]', regex=True)
# df2['update_text'] = df2['update_text'].astype(str)
print(df2['update_text'].apply(type).unique())
```

```python
for i in df2['update_text']:
  print(i)
```

```python
# # –ó–∞–º–µ–Ω–∞ –ø—É—Å—Ç—ã—Ö —Å—Ç—Ä–æ–∫ –∏ NaN
df2['update_text'] = df2['update_text'].fillna('[–ø—É—Å—Ç–æ]')
df2['update_text'] = df2['update_text'].replace('', '[–ø—É—Å—Ç–æ]')

# # –ù–∞—Å—Ç—Ä–æ–π–∫–∞ –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ç–æ—Ä–∞
vectorizer = CountVectorizer(
     min_df=1,          # –£—á–∏—Ç—ã–≤–∞—Ç—å –≤—Å–µ —Å–ª–æ–≤–∞
    stop_words=None,   # –ù–µ —É–¥–∞–ª—è—Ç—å —Å—Ç–æ–ø-—Å–ª–æ–≤–∞
     token_pattern=r'(?u)\b\w+\b'  # –í–∫–ª—é—á–∞—Ç—å —Ü–∏—Ñ—Ä—ã
)
# print("–°–ª–æ–≤–∞—Ä—å:", vectorizer.get_feature_names_out())
```

```python
X = vectorizer.fit_transform(df2['update_text']).toarray()
# print("–ú–∞—Ç—Ä–∏—Ü–∞:", X)
print("–†–∞–∑–º–µ—Ä–Ω–æ—Å—Ç—å X:", X.shape)
```

```python
# –ü–æ–ª—É—á–∏—Ç—å —Å–ø–∏—Å–æ–∫ –≤—Å–µ—Ö —Å–ª–æ–≤ (–ø—Ä–∏–∑–Ω–∞–∫–æ–≤)
feature_names = vectorizer.get_feature_names_out()
print("–°–ª–æ–≤–∞—Ä—å –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:", feature_names)
# –ü—Ä–æ–π—Ç–∏—Å—å –ø–æ –∫–∞–∂–¥–æ–º—É –¥–æ–∫—É–º–µ–Ω—Ç—É –∏ –µ–≥–æ –≤–µ–∫—Ç–æ—Ä—É
for doc_idx, doc_vector in enumerate(X):
    print(f"\n–î–æ–∫—É–º–µ–Ω—Ç {doc_idx}:")
    # –ü—Ä–æ–π—Ç–∏—Å—å –ø–æ –∫–∞–∂–¥–æ–º—É –ø—Ä–∏–∑–Ω–∞–∫—É (—Å–ª–æ–≤—É) –≤ –¥–æ–∫—É–º–µ–Ω—Ç–µ
    for word_idx, count in enumerate(doc_vector):
        if count > 0:  # –ü–æ–∫–∞–∑–∞—Ç—å —Ç–æ–ª—å–∫–æ –Ω–µ–Ω—É–ª–µ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏
            word = feature_names[word_idx]
            print(f"–°–ª–æ–≤–æ '{word}': {count} —Ä–∞–∑")
```

```python
import numpy as np
import pandas as pd
from nltk.tokenize import word_tokenize
import nltk
from pymorphy3 import MorphAnalyzer
from nltk.stem.snowball import RussianStemmer
from string import punctuation
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# –æ–¥–∏–Ω —Ä–∞–∑ –∂–∞—Ç—å
from nltk import word_tokenize

df2["update_text"] = df2["update_text"].apply(
    lambda x: word_tokenize(x) if isinstance(x, str) and x.strip() != '' else []
)
```

```python
df2
```

```python
#–≤–µ–∫—Ç–æ—Ä–∞ –∫–∞–∂–¥–æ–≥–æ —Å–ª–æ–≤–∞ –∏–∑ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–π
preprocessing_text = list()
for i in df2["update_text"]:
  preprocessing_text.append(i)
```

```python
vectors_word = []
for i in preprocessing_text:
  words = []
  for j in i:
    if j in vocab:
    #vocab.index(i) - –Ω–∞—Ö–æ–¥–∏—Ç –∏–¥–µ–∫—Å –≤ vocab
    #list(wv.index_to_key)[vocab.index(i)] - –≤—ã–≤–æ–¥–∏—Ç —Å–ª–æ–≤–æ —Å –µ–≥–æ —á–∞—Å—Ç—å—é —Ä–µ—á–∏
    #wv[list(wv.index_to_key)[vocab.index(i)]] - –ø–æ–ª—É—á–∞–µ–º –≤–µ–∫—Ç–æ—Ä
      words.append(wv[list(wv.index_to_key)[vocab.index(j)]])
    else:
      vectors_word.append(list(np.zeros(300)))
  vectors_word.append(words)
vector_text = np.mean(vectors_word, axis = 0)
```

```python
#–≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—è –∫–∞–∂–¥–æ–≥–æ —Å–ª–æ–≤–∞ –ø–æ –æ—Ç–¥–µ–ª—å–Ω–æ—Å—Ç–∏
print(vectors_word)
```
